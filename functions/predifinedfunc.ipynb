importing libraries 
- loading the modules 
`import module`
- loading all items from the modules 
`from module import *`
- loading specific func from modules
`from module import func`
- loading specific func from modules with alas
`from module import func as alias`
- loading modules as alias 
`import module as alias`

```python
random.seed(42)
print(random.random())
print(random.randint(1,10))
nums = [1,2,3,4,5,6,7,8,9]
random.shuffle(nums)
print(nums)
print(random.choice(nums))
print(random.choices(nums,k=3))
print(random.sample(nums,k=3))
```
