```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
t = np.linspace(0, 3*np.pi, 50)
x = np.sin(t)
y = np.cos(t)
```

```python
plt.figure(figsize=(15,7))
plt.plot(t, x, 'ro--') # 0 - 6.8 range 
plt.plot(t, y, 'h-', markersize =10 )
```
