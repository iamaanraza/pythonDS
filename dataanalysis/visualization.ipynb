```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
t = np.linspace(0, 3*np.pi, 50)
x = np.sin(t)
y = np.cos(t)
```

```python
plt.figure(figsize=(15,7))
plt.plot(t, x, 'ro--') # 0 - 6.8 range 
plt.plot(t, y, 'h-', markersize =10 )
```

```python
plt.figure(figsize=(15,7))
plt.plot(t,x, marker='o',
         markersize=20,markeredgecolor='blue',
         markerfacecolor='yellow')
plt.plot(t,y,'h',markersize=10)
```

```python
years = [2010,2011,2012,2013,2014,2015]
prices = [120,90,140,150,180,130]
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0,0,.2,.2])
ax3 = fig.add_axes([.8,.8,.2,.2]) # X,Y, width , height
ax4 = fig.add_axes([.8,0,.2,.2])
ax5 = fig.add_axes([.4,.3,.3,.3])
```

```python
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_axes([0,0,.8,.8])
ax1.bar(years, prices) # for graph 
ax1.pie(prices, labels = years, autopct='%.2f%%') # for pie 
plt.show() # remove extra info.
```

# Customization

```python
plt.figure(figsize=(6,6), facecolor = 'lightblue')
plt.plot(years, prices, 'ro--')
plt.title("prices of stocks")
plt.hlines([140,140,150],[2010,2012,2010] ,[2015,2015,2013],
            linestyles=['dashed','solid'])
plt.vlines([2012,2014],[min(prices),min(prices)],[180,180],
           linestyles=['dashed','solid'],colors=['green','blue'],
           label='prices')
plt.text(2011, 144,'Important year', fontdict={'size':10})
plt.text(2012.3, 130, '\n in price')
plt.xlabel("Years")
plt.ylabel("Prices")
plt.show()
```
